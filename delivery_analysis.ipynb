```python
for categ, group in df_orders.groupby("product_category"):
    group = group.dropna(subset=["valid_delivered_qty"])
    qty = group["qty"]
    qty_sum = qty.sum()
    no_valid_qty = (qty - group["valid_delivered_qty"])
    no_valid_qty_sum = (no_valid_qty > 0).sum()
    print(
        f"category {categ}: \nколичество недошедшего товара:",
        f"{no_valid_qty.sum()}({(no_valid_qty.sum() * 100) / qty_sum:.2f}%)",
        f"\nколичество не дошедших заказов {no_valid_qty_sum}({no_valid_qty_sum * 100 / qty.count():.2f}%)",
    )
```
